```python
# Step 1: Importing Libraries
import pandas as pd
```

```python
# Step 2: Importing Dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
df = pd.read_csv(url, sep = '|')
```

```python
# Step 3: Getting to know your data
df.head()
```

```python
# Discover what is the mean age per occupation
round(df.groupby('occupation')['age'].mean(), 2)
```

```python
df['gender'] = df['gender'].apply(lambda gender: 1 if gender == 'M' else 0)
```

```python
# Discover the male ratio per occupation and sort it from the most to the least
(df.groupby('occupation')['gender'].sum()/df['occupation'].value_counts().sort_index(ascending = True))
```

```python
# For each occupation, calculate the minimum and maximum ages
df.groupby('occupation')['age'].agg(['min', 'max'])
```

```python
# For each combination of occupation and gender, calculate the mean age
df.groupby(['occupation', 'gender'])['age'].mean()
```

```python
# For each occupation present the percentage of women and men
(df.groupby('occupation')['gender'].sum() / df['occupation'].value_counts().sort_index(ascending = True))*100
```

```python
# For each occupation present the percentage of women and men
(df.groupby(['occupation', 'gender'])['age'].agg('count') / df.groupby('occupation')['age'].agg('count'))*100
```
